**HEALTH RECOMMENDATION CHATBOT**. 

```python
import nltk
from nltk.chat.util import Chat, reflections

# Download required NLTK data
nltk.download('punkt')

# Define patterns and responses for health advice
pairs = [
    (r"(.*) headache(.*)", ["It seems you might be experiencing a headache. Try to rest, drink plenty of water, and avoid screen time. If symptoms persist, consult a doctor."]),
    (r"(.*) fever(.*)", ["A fever could indicate an infection. Make sure to stay hydrated and rest. If it lasts more than 48 hours or is accompanied by severe symptoms, seek medical advice."]),
    (r"(.*) cold(.*)", ["For a common cold, stay warm, hydrate well, and get plenty of rest. Consider using over-the-counter remedies for relief, but if symptoms persist, contact your healthcare provider."]),
    (r"(.*) cough(.*)", ["Coughing can be due to many reasons. Stay hydrated and avoid irritants like smoke. If the cough lasts more than a few days or is severe, see a doctor."]),
    (r"(.*) fatigue(.*)", ["Fatigue can result from many things. Make sure you're sleeping enough, eating well, and managing stress. If itâ€™s persistent, you should see a doctor to rule out any underlying issues."]),
    (r"(.*) (stress|anxiety)(.*)", ["Stress and anxiety can affect both mental and physical health. Practice relaxation techniques like deep breathing, meditation, and ensure you're getting enough sleep."]),
    (r"(.*) (exercise|workout)(.*)", ["Regular exercise is great for overall health. Aim for at least 30 minutes of moderate activity most days of the week. Always consult with your healthcare provider before starting a new workout routine."]),
    (r"(.*) (diet|nutrition)(.*)", ["A balanced diet is key to good health. Make sure to include a variety of fruits, vegetables, whole grains, and lean proteins. Drink plenty of water and avoid excessive sugar."]),
    (r"(.*) sleep(.*)", ["Quality sleep is essential for good health. Aim for 7-9 hours of sleep each night. Try to maintain a consistent sleep schedule and avoid caffeine or screens before bed."]),
    (r"quit", ["Goodbye! Stay healthy and take care."]),
    (r"(.*)", ["I can only provide general health recommendations. If you have specific concerns, please consult a healthcare provider."]),
]

# Create the chatbot instance
health_chatbot = Chat(pairs, reflections)

# Start the chatbot conversation loop
print("HealthBot: Hello! I can provide health recommendations based on your symptoms. Type 'quit' to exit.")
while True:
    user_input = input("You: ")

    # Exit the conversation if the user types 'quit'
    if user_input.lower() == "quit":
        print("HealthBot: Goodbye! Stay healthy and take care.")
        break
    
    # Get and print the chatbot's response
    response = health_chatbot.respond(user_input)
    print(f"HealthBot: {response}")
```
